```python
import pandas as pd
import gmaps
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
file = "sign_data.csv"
sign_data = pd.read_csv(file, encoding="ISO-8859-1")
sign_data.head()
```

```python
sign_data.columns
```

```python
organized_sign_data = sign_data[["X", "Y", "SIGNNAME", "Signal", "VISIBILITY", "SIGNCONDITION", "XCOORD", "YCOORD"]]
organized_sign_data.head()
```

```python
locations = organized_sign_data[["Y", "X"]]
fig = gmaps.figure()
markers = gmaps.marker_layer(locations)
fig.add_layer(markers)
fig
```
