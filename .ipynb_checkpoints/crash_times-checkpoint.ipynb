```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = "Crashes_in_DC (1).csv"
crash_data = pd.read_csv(file, encoding="ISO-8859-1")
crash_data.head()
```

```python
crash_time = crash_data[["ADDRESS", "REPORTDATE"]]
crash_time.head()
```

```python
crash_time['DATE'], crash_time["FULLTIME"]= crash_data['REPORTDATE'].str.split('T', 1).str
crash_time.head()
```

```python
crash_time["TIME"], crash_time["JUNK"] = crash_time["FULLTIME"].str.split(".", 1).str
crash_time.head()
```

```python
crash_time["HOUR"], crash_time["MINUTE"] = crash_time["TIME"].str.split(":", 1).str
crash_time.head()
```

```python
pd.to_numeric(crash_time['HOUR'], errors='coerce')
crash_time.head()
```

```python
bins = [-1, 6, 9, 15, 18, 24]
time_frames = ["Midnight-6:00", "Morning Rush Hour", "9:00-15:00", "Afternoon Rush Hour", "18:00-Midnight"]

crash_time["TIME DEMOGRAPHIC"] = pd.cut(pd.to_numeric(crash_time['HOUR'], errors='coerce'), bins, labels=time_frames)
crash_time
```

```python
time_demo = crash_time.groupby(["TIME DEMOGRAPHIC"])
crash_count = time_demo["ADDRESS"].count()

demographic_df = pd.DataFrame({"Crash Count": crash_count})
demographic_df
```

```python
labels = ["Midnight-6:00", "Morning Rush Hour", "9:00-15:00", "Afternoon Rush Hour", "18:00-Midnight"]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue", "purple"]
sizes = crash_count
explode = explode = (0.1, 0, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=50)
plt.axis("equal")
plt.title("Times of Crashes in DC")
plt.savefig("Crash_Times.png")
plt.show()
```
