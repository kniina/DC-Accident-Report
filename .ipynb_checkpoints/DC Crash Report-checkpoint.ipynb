```python
url = "https://maps2.dcgis.dc.gov/dcgis/rest/services/DCGIS_DATA/Public_Safety_WebMercator/MapServer/24/query?where=1%3D1&outFields=*&outSR=4326&f=json"
```

```python
response = requests.get(url).json()
```

```python
data = response['features']
crash_data = []
```

```python
for t in range(185):
    i = 0
    for x in data:
        crash_data.append(data[i]['attributes'])
        i = i + 1
```

```python
master_df = pd.DataFrame(crash_data)
master_df.head()

```

```python
#Create bike only dataframe
bike_df = master_df[["ADDRESS", "CRIMEID", "FATAL_BICYCLIST", "FROMDATE", "LATITUDE", "LONGITUDE", "MAJORINJURIES_BICYCLIST", "MINORINJURIES_BICYCLIST", "XCOORD", "YCOORD"]]
bike_df_reduced = bike_df.loc[(bike_df["MAJORINJURIES_BICYCLIST"] >= 1) | (bike_df["MINORINJURIES_BICYCLIST"]>= 1) | (bike_df["FATAL_BICYCLIST"]>= 1), :]
bike_df["MAJORINJURIES_BICYCLIST"].sum()
```

```python
#Create vehicle only dataframe
vehicle_df = master_df[["ADDRESS", "CRIMEID", "FATAL_DRIVER", "FROMDATE", "LATITUDE", "LONGITUDE", "MAJORINJURIES_DRIVER", "MINORINJURIES_DRIVER", "XCOORD", "YCOORD"]]
vehicle_df_reduced = vehicle_df.loc[(vehicle_df["MAJORINJURIES_DRIVER"] >= 1) | (vehicle_df["MINORINJURIES_DRIVER"]>= 1) | (vehicle_df["FATAL_DRIVER"]>= 1), :]
vehicle_df.head()
```

```python
#Create pedestrian only dataframe
pedest_df = master_df[["ADDRESS", "CRIMEID", "FATAL_PEDESTRIAN", "FROMDATE", "LATITUDE", "LONGITUDE", "MAJORINJURIES_PEDESTRIAN", "MINORINJURIES_PEDESTRIAN", "XCOORD", "YCOORD"]]

pedest_df_reduced = pedest_df.loc[(pedest_df["MAJORINJURIES_PEDESTRIAN"] >= 1) | (pedest_df["MINORINJURIES_PEDESTRIAN"] >= 1) | (pedest_df["FATAL_PEDESTRIAN"] >= 1),:]
pedest_df.head()
```

```python
pedest_df.dtypes
```

```python
#Convert 'FROMDATE' column into date format
bike_df["FROMDATE"]=pd.to_datetime(bike_df['FROMDATE'],unit='ms')

vehicle_df["FROMDATE"]=pd.to_datetime(vehicle_df["FROMDATE"],unit='ms')

pedest_df["FROMDATE"]=pd.to_datetime(pedest_df['FROMDATE'],unit='ms')
bike_df
```

```python
#List only the years for bike data in the last column

bike_df["YEAR"]=pd.DatetimeIndex(bike_df["FROMDATE"]).year
bike_df
```

```python
#List only the years for vehicle data in the last column

vehicle_df["YEAR"]=pd.DatetimeIndex(vehicle_df["FROMDATE"]).year
vehicle_df.head()
```

```python
#List only the years for pedestrian data in the last column

pedest_df["YEAR"]=pd.DatetimeIndex(pedest_df["FROMDATE"]).year
pedest_df
```

```python
#aggregate table of #of minor and major injuries + fatalities
bike_agg = bike_df_reduced.groupby(["YEAR"])[['MINORINJURIES_BICYCLIST','MAJORINJURIES_BICYCLIST','FATAL_BICYCLIST']].sum()
bike_agg
```

```python
#aggregate table of #of minor and major injuries + fatalities
vehicle_agg = vehicle_df_reduced.groupby(["YEAR"])[['MINORINJURIES_DRIVER','MAJORINJURIES_DRIVER','FATAL_DRIVER']].sum()
vehicle_agg
```

```python
#aggregate table of #of minor and major injuries + fatalities
pedest_agg = pedest_df_reduced.groupby(["YEAR"])[['MINORINJURIES_PEDESTRIAN','MAJORINJURIES_PEDESTRIAN','FATAL_PEDESTRIAN']].sum()
pedest_agg
```
